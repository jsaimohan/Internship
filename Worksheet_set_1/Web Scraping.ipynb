```python
#question:2

page=requests.get('https://www.imdb.com/list/ls091520106/?sort=list_order,asc&st_dt=&mode=simple&page=1&ref_=ttls_vw_smp')
page

soup = BeautifulSoup(page.content)
soup

titles = []

for i in soup.find_all('div', class_='col-title'):
    titles.append(i.text.replace('\n',''))
    
titles

rating = []

for i in soup.find_all('div', class_='col-imdb-rating'):
    rating.append(i.text.split()[0])
    
rating

year = []

for i in soup.find_all('span', class_='lister-item-year text-muted unbold'):
    year.append(i.text.split()[0])
    
year

print(len(titles),len(rating),len(year))

import pandas as pd
df = pd.DataFrame({'Titles':titles,'Rating':rating,'Year':year})
df
```

```python
#Question:4
page=requests.get('https://meesho.com/bags-ladies/pl/p7vbp')
page
soup = BeautifulSoup(page.content)
soup
titles = []

for i in soup.find_all('p', class_='Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS'):
    titles.append(i.text.replace('\n',''))
    
titles
Price = []

for i in soup.find_all('div', class_='Card__BaseCard-sc-b3n78k-0 iLPHgK NewProductCard__PriceRow-sc-j0e7tu-5 eyya-Dr NewProductCard__PriceRow-sc-j0e7tu-5 eyya-Dr'):
    Price.append(i.text.split('₹')[1])
    
Price
Discount = []

for i in soup.find_all('span', class_='Text__StyledText-sc-oo0kvp-0 lnonyH'):
    Discount.append(i.text)
    
Discount
print(len(titles),len(Price),len(Discount))
import pandas as pd
df = pd.DataFrame({'Titles':titles,'Price':Price,'Discount':Discount})
df
```

```python
#question:7
page=requests.get('https://coreyms.com/')
page
soup = BeautifulSoup(page.content)
soup
Title = []

for i in soup.find_all('h2', class_='entry-title'):
    Title.append(i.text)
Title=Title[0:4]+ Title[5:]
Title
time= []

for i in soup.find_all('time', class_='entry-time'):
    time.append(i.text) 
time=time[0:4]+time[5:] 
time
content = []


for i in soup.find_all('div', class_='entry-content'):
    content.append(i.text.replace('\n', ''))
    
content=content[0:4]+content[5:]
content
link = []


for i in soup.find_all('iframe', class_='youtube-player'):
    link.append(i['src'])
    
link
print(len(Title),len(time),len(content),len(link))
import pandas as pd
df = pd.DataFrame({'Title':Title,'Time':time,'Content':content,'Link':link})
df
```

```python
#question:9
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
soup = BeautifulSoup(page.content)
soup
Title = []

for i in soup.find_all('div', class_='restnt-info cursor'):
    Title.append(i.text)
    
Title

cuisine = []

for i in soup.find_all('span', class_='double-line-ellipsis'):
    cuisine.append(i.text.split()[6:])
    
cuisine
Loc = []

for i in soup.find_all('div', class_='restnt-loc ellipsis'):
    Loc.append(i.text)
    
Loc
rating = []

for i in soup.find_all('div', class_='restnt-rating rating-4'):
    rating.append(i.text)
    
rating
image = []

for i in soup.find_all('img', class_='no-img'):
    image.append(i['data-src'])
    
image
print(len(Title),len(cuisine),len(Loc),len(rating),len(image))
import pandas as pd
df = pd.DataFrame({'Title':Title,'cuisine':cuisine,'Loc':Loc,'rating':rating,'image':image})
df
```
