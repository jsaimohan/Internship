```python
driver=webdriver.Chrome('chromedriver.exe')
#Question:1
url='http://www.naukri.com/'
driver.get(url)
search_job= driver.find_element_by_class_name("suggestor-input ")
search_job.send_keys('Data Analyst')

search_locn=driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input')
search_locn.send_keys('Bangalore')
search_btn=driver.find_element_by_xpath("/html/body/div/div[2]/div[3]/div/div/div[6]")
search_btn
search_btn.click()

#extracting titles
titles_tags=driver.find_elements_by_xpath('//a[@class="title fw500 ellipsis"]')
titles_tags
len(titles_tags)
titles_tags[0:2]

job_titles=[]

for i in titles_tags:
    job_titles.append(i.text)
job_titles[:10]

#extracting company names
company_tag=driver.find_elements_by_xpath('//a[@class="subTitle ellipsis fleft"]')
company_tag
company_tag[0:2]
company_tags=[]

for i in company_tag:
    company_tags.append(i.text)
company_tags[:10]

#extracting experience required
exp_tags=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi experience"]')
exp_tags[:10]
Exp_tags=[]

for i in exp_tags:
    Exp_tags.append(i.text)
Exp_tags[0:10]

#extracting job location
job_loc=driver.find_elements_by_xpath('//li[@class="fleft grey-text br2 placeHolderLi location"]')
job_loc
len(job_loc)
job_loc[0:2]

Loc_tags=[]

for i in job_loc:
    Loc_tags.append(i.text)
Loc_tags[:10]
print(len(job_titles),len(company_tags),len(Exp_tags),len(Loc_tags))

jobs=pd.DataFrame()
jobs['Job-title']=job_titles
jobs['Job-location']=Loc_tags
jobs['Company_name']=company_tags
jobs['Experience_required']=Exp_tags
jobs[:10]
```
